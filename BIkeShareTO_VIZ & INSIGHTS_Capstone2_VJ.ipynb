#### ANALYSIS OF PERCENTAGE CHANGE IN RIDERSHIP IN LAST 3 YEARS (2020,2019 and 2018)

* From the below Visualizations, it is clear that the ridership decreased by 117% in April 2020 and by 40 % in May 2020 when comapred to April & May 2019 respectively. This decrease is due to the initial face of COVID lockdown.
* Though an increasing trend is seen after the ridership bounced back in June 2020 with a positve % change, it is to be noted that the ridership in July & August 2020 have increased only by 14% and 7.8% respectivley which is less than the percentage change from previous years.
* Pre COVID time (Q1 2020) has seen a postive percentage change when compared to previous years.

### CONCLUSION 

* A detailed exploratory analysis on the ridership for years 2018 till 2020 reavealed that the ridership is on a steady rise since 2018 with the exception being the COVID lockdown time *(Week 11 till week 21)*.
* The Pre COVID lockdown time and cold Quater(Q1 2020) has infact seen a positive percentage change when compared previous 2 years. 
* With respect to the User type ( Casual and registered Members), Casual riders have drastically increased in Q3 2020 as more people want to be active to keep healthy. 
* The highest ridership day in 3 years was May 23, 2020 with 14,020 rides reason being the first warm day of 2020 post Covid-19 lockdown.
* The days which had all time least ridership is only due to Covid-19 lockdown and cold temperatures, which are out of scope for manual intervention to increase ridership.
